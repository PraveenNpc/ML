```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m = 0
    c = 0
    iterations = 100
    learning_rate = 0.05
    n= len(x)
    cost_list=([])
    for i in range (iterations):
        y_predicted = m*x + c
        cost = (1/n)* sum([val**2 for val in (y-y_predicted)])
        dm = - (2/n)* sum (x*(y-y_predicted))
        dc = - (2/n)* sum (y-y_predicted)
        m = m - learning_rate * dm
        c = c - learning_rate * dc
        print ("iter= {}, m= {},  c= {}, cost= {}".format(i , m , c , cost))
        print("-----"*20)
        cost_list.append(cost)
    print (cost_list)
    return(cost_list,m,c,cost)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])
print(x)
print(y)
```

```python
(cost_list,m,c,cost) = gradient_descent (x,y)
```

```python
print(cost_list)
```

```python
print("slope m = {}, intercept c = {}, cost ={} ".format(m , c , cost))
```
